# Plotly, matplotlib, and mplexporter: Annotations

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import matplotlib.cm as cm
from mplexporter.renderers import fig_to_plotly
```

```python
gs = gridspec.GridSpec(2, 1)
fig = plt.figure()
ax1 = fig.add_subplot(gs[0,0])
x = np.arange(0, 3, 0.01)*2*np.pi
num_lines = 20
colors = cm.rainbow(np.linspace(0, 1, num_lines))
for iii in range(0, num_lines):
    y = x
    for jjj in range(1, iii+1):
        sign = (-1)**jjj
        deg = 1+2*jjj
        y = np.add(y, sign*x**(deg)/np.math.factorial(deg))
    ct = 0
    for amp in y:
        ct += 1
        if abs(amp) > 10:
            break
    x_plot = x[:ct]
    y_plot = y[:ct]
    ax1.plot(x_plot, y_plot, color=colors[iii], label='L{}'.format(iii))
    ax1.text(x_plot[-15], y_plot[-15], '{}'.format(1+2*iii), fontsize=6)
ax1.set_ylim(-10,10)
ax1.set_ylabel('y(x)')
ax1.set_title('Approximation to sin(x)', size=14)
ax1.text(0.01, 0.99, 'num=deg', transform=ax1.transAxes, ha='left', va='top', fontsize=12)

ax2 = fig.add_subplot(gs[1,0])
x = np.arange(-2, 5, 0.01)
num_lines = 20
colors = cm.rainbow(np.linspace(0, 1, num_lines))
for iii in range(0, num_lines):
    y = x
    for jjj in range(1, iii+1):
        sign = (-1)**jjj
        deg = 1+jjj
        y = np.add(y, sign*x**(deg)/deg)
    ct = int(len(x)/2)
    while abs(y[ct]) < 3:
        ct += 1
    x_plot = x[:ct]
    y_plot = y[:ct]
    ax2.plot(x_plot, y_plot, color=colors[iii], label='L{}'.format(iii))
    ax2.text(x_plot[-28], y_plot[-28], '{}'.format(1+iii), fontsize=6)
ax2.set_ylim(-3,3)
ax2.set_xlim(-2,4)
ax2.set_xlabel('x')
ax2.set_ylabel('y(x)')
ax2.set_title('Approximation to ln(1 + x)', size=14)
ax2.text(0.01, 0.99, 'num=deg', transform=ax2.transAxes, ha='left', va='top', fontsize=12)
gs.update(wspace=0.2, hspace=0.4)
```
